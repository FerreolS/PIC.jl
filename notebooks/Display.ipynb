## Statistics of resulting parameters

```julia
a_csv = CSV.read("atab.csv", DataFrame; datarow = 1, transpose = true);
a0_hist = (a_csv[:,2]);
a1_hist = (a_csv[:,3]);
a2_hist = (a_csv[:,4]);

fwhm_csv = CSV.read("fwhmtab.csv", DataFrame; datarow = 1, transpose=true);
fwhm0_hist = (fwhm_csv[:,2]);
fwhm1_hist = (fwhm_csv[:,3]);
fwhm2_hist = (fwhm_csv[:,4]);

cx_csv = CSV.read("cxtab.csv", DataFrame; datarow = 1, transpose=true);
cx1_hist = (cx_csv[:,3]);
cx2_hist = (cx_csv[:,4]);

cy_csv = CSV.read("cytab.csv", DataFrame; datarow = 1, transpose=true);
cy1_hist = (cy_csv[:,3]);
cy2_hist = (cy_csv[:,4]);
```

```julia
cx1 = (cx1_hist .- median(cx1_hist))/mad(cx1_hist);
cx2 = (cx2_hist .- median(cx2_hist))/mad(cx2_hist);
cy1 = (cy1_hist .- median(cy1_hist))/mad(cy1_hist);
cy2 = (cy2_hist .- median(cy2_hist))/mad(cy2_hist);
fwhm0 = (fwhm0_hist .- median(fwhm0_hist)/mad(fwhm0_hist));
fwhm1 = (fwhm1_hist .- median(fwhm1_hist)/mad(fwhm1_hist));
fwhm2 = (fwhm2_hist .- median(fwhm2_hist)/mad(fwhm2_hist));

#Pour detecter les "bad lenslets", on regarde la distance au centre de la lenslet
x= sqrt.((cx1 .+ cx2 .+ cy1 .+ cy2) .^2);
```

```julia
N = 100
for i:N
    if  x[i,:] .> 5
        println("bad lenslet")
    end
end
```

```julia
#amplitudes
h1 = histogram([a0_hist a1_hist a2_hist], title = "amplitudes", label = ["a0" "a1" "a2"],
    bins = LinRange(-200,1500, 100), xlabel = "position", ylabel = "amplitude");

#fwhm
h2 = histogram([fwhm0 fwhm1 fwhm2], title = "fwhm", label = ["fwhm0" "fwhm1" "fwhm2"],
    bins = LinRange(-40,-10, 100),  xlabel = "position", ylabel = "fwhm", legend = :best);

#polynomial coefficients cx
h3 = histogram([cx1 cx2], title = "cx", label = ["cx1" "cx2"],
    bins = LinRange(-5,5,100), xlabel = "position", ylabel = "cx", legend = :best);

#polynomial coefficients cy
h4 = histogram([cy1 cy2], bins = LinRange(-5,5,100),
      label = ["cy1" "cy2"], xlabel = "position", ylabel = "cy", title = "cy", legend = :best, 
      colour = ["orange" "lightgreen"]);
```
