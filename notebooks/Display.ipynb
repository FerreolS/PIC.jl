```julia
#Packages
using PIC

using Plots, StatsBase, Statistics, TwoDimensional, OptimPackNextGen, DelimitedFiles
using CSV, DataFrames

plotly()
```

## Statistics of resulting parameters

```julia
plotly(size=(600,400))
#amplitudes
a_csv = CSV.read("atab.csv", DataFrame; datarow = 1, transpose = true);
a0_hist = (a_csv[:,2]);
a1_hist = (a_csv[:,3]);
a2_hist = (a_csv[:,4]);
h1 = histogram([a0_hist a1_hist a2_hist], title = "amplitudes", label = ["a0" "a1" "a2"],
    bins = LinRange(-200,1500, 100), xlabel = "position", ylabel = "amplitude")
```

```julia
plotly(size=(600,400))
#fwhm
fwhm_csv = CSV.read("fwhmtab.csv", DataFrame; datarow = 1, transpose=true);
fwhm0_hist = (fwhm_csv[:,2]);
fwhm1_hist = (fwhm_csv[:,3]);
fwhm2_hist = (fwhm_csv[:,4]);
h2 = histogram([fwhm0_hist fwhm1_hist fwhm2_hist], title = "fwhm", label = ["fwhm0" "fwhm1" "fwhm2"],
    bins = LinRange(1.5,3.5, 100),  xlabel = "position", ylabel = "fwhm", legend = :best)
```

```julia
plotly(size=(600,400));
#polynomial coefficients cx
cx_csv = CSV.read("cxtab.csv", DataFrame; datarow = 1, transpose=true);
cx1_hist = (cx_csv[:,3]);
cx2_hist = (cx_csv[:,4]);
h3 = histogram([cx1_hist cx2_hist], title = "cx", label = ["cx1" "cx2"],
    bins = LinRange(-20,20,100), xlabel = "position", ylabel = "cx", legend = :best);

plot(h3)
```

Comparaison aux cx initiaux: plus etale, maximum pas en 0, amplitude maximale plus faible

```julia
plotly(size=(600,400));
#polynomial coefficients cy
cy_csv = CSV.read("cytab.csv", DataFrame; datarow = 1, transpose=true);
cy1_hist = (cy_csv[:,3]);
cy2_hist = (cy_csv[:,4]);
h4 = histogram([cy1_hist cy2_hist], bins = LinRange(-100,130,200),
     label = ["cy1" "cy2"], xlabel = "position", ylabel = "cy", title = "cy", legend = :best, 
     colour = ["orange" "lightgreen"]);

plot(h4)
```

Comparaison au cy initiaux: comportement tres similaire

```julia
plotly(size=(900,300));
#polynomial coefficients cy1 and cy2
h5 = histogram(cy1_hist, label = "cy1",
     bins = LinRange(100,105,100), ylabel = "cy1", legend = :best, color = "orange");

h6 = histogram(cy2_hist, label = "cy2",
     bins = LinRange(-100,-25,100), ylabel = "cy2", legend = :best, color = "lightgreen");

plot(h5, h6)
```

Comparaison cy1 et cy2 initiaux: amplitudes maximales plus faible, divisee par 2 pour cy2.Pour cy1 nombre d'outliers diminues (courbe plus lisse), mais toujours le meme comportement asymetrique (non gaussien).
